Das Script liest Datein der Art "Analytics Alle Websitedaten Audience Overview 20220801-20220831 20210801-20210831" ein,<br>
wandelt das Datum ins ISO-Format um, verk√ºrzt 'Date Range', ersetzt den Namen durch 'Year' und sortiert nach dem Datum

```python
import pandas as pd
import numpy as np
```

```python
def date_range(datrngstr):
    if datrngstr == 'Aug 1, 2021 - Aug 31, 2021':
        return 2021
    elif datrngstr == 'Aug 1, 2022 - Aug 31, 2022':
        return 2022
    else:
        return 0
```

```python
csv_data_file = 'data.csv'
cols = ['Day Index','Date Range','Users']
df = pd.read_csv(csv_data_file,sep=',',header=0, usecols=cols, parse_dates=[0], index_col=0,thousands=',')
```

```python
df['Date Range'] = df['Date Range'].apply(lambda x: date_range(x))
df['Users'] = df['Users'].astype('int')
df.rename(columns={'Date Range':'Year'}, inplace=True)
```

```python
df[pd.notnull(df.index)].sort_index()  # pd.notnull(df.index) filtert NaT-Werte heraus
```

```python
pivot = df.pivot_table(index = ['Year'],
      values = ['Users'],
      aggfunc = 'sum', fill_value=0, margins=True)
```
